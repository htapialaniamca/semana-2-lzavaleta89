```python
import numpy as np
import pandas as pd
import matplotlib as plt
import ipyparallel as ipp
%matplotlib inline
```

```python
#Obtenemos los ID de clientes
cliente = ipp.Client()
cliente.ids
```

```python
#Asignamos los clientes a variable
dview = cliente[:]
```

```python
#Funci√≥n a probar
def fun1(x):
    import numpy as np
    y = 2**x
    for i in np.arange(int(y**0.5)+1):
        y%i
```

```python
import timeit 

#Arreglos para almacenar los tiempos obtenidos tras ejecutar la funci√≥n map_sync
arregloT1 = []

for i in np.arange(25) : #se probaran con los primeros 25 elementos
    ti = timeit.default_timer()
    dview.map_sync(fun1, range(25))
    arregloT1.append(timeit.default_timer() - ti)
    
import matplotlib.pyplot as plt
plt.plot(arregloT1, label = "Resultado obtenidos con map_sync")
plt.legend(loc="best")
plt.show()
```

```python
import timeit 

#Arreglos para almacenar los tiempos obtenidos tras ejecutar la funci√≥n apply_sync
arregloT2 = []

for i in np.arange(25) : #se probaran con los primeros 25 elementos
    ti = timeit.default_timer()
    dview.apply_sync(fun1, 2**i)
    arregloT2.append(timeit.default_timer() - ti)
    
import matplotlib.pyplot as plt
plt.plot(arregloT2, label = "Resultado obtenidos con apply_sync")
plt.legend(loc="best")
plt.show()
```

```python
#Diferencia de tiempos
import matplotlib.pyplot as plt
plt.plot(arregloT1, label = "Resultado obtenidos con map_sync")
plt.plot(arregloT2, label = "Resultado obtenidos con apply_sync")
plt.legend(loc="best")
plt.show()
```

## <font color='blue'>Funci√≥n primosQA</font>

```python
def primosQA(x):
    import numpy as np
    return all(x % i for i in np.arange(int(x**0.5)+1)[2:])
```

```python
import timeit 

qaT1 = []

for i in np.arange(25) : #se probaran con los primeros 25 elementos
    ti = timeit.default_timer()
    dview.map_sync(primosQA, range(25))
    qaT1.append(timeit.default_timer() - ti)
    
import matplotlib.pyplot as plt
plt.plot(qaT1, label = "Resultado obtenidos con map_sync")
plt.legend(loc="best")
plt.show()
```

```python
import timeit 

qaT2 = []

for i in np.arange(25) : #se probaran con los primeros 25 elementos
    ti = timeit.default_timer()
    dview.apply_sync(primosQA, 2**i)
    qaT2.append(timeit.default_timer() - ti)
    
import matplotlib.pyplot as plt
plt.plot(qaT2, label = "Resultado obtenidos con apply_sync")
plt.legend(loc="best")
plt.show()
```

### Comparando tiempos de primosQA

```python
#Diferencia de tiempos
import matplotlib.pyplot as plt
plt.plot(qaT1, label = "Resultado obtenidos con map_sync")
plt.plot(qaT2, label = "Resultado obtenidos con apply_sync")
plt.legend(loc="best")
plt.show()
```

## <font color='blue'>Funci√≥n primosQB</font>

```python
def primosQB(x):
    import numpy as np
    return np.all(np.array([x % i for i in np.arange(int(x**0.5)+1)[2:]]))
```

```python
import timeit 

qbT1 = []

for i in np.arange(25) : #se probaran con los primeros 25 elementos
    ti = timeit.default_timer()
    dview.map_sync(primosQB, range(25))
    qbT1.append(timeit.default_timer() - ti)
    
import matplotlib.pyplot as plt
plt.plot(qbT1, label = "Resultado obtenidos con map_sync")
plt.legend(loc="best")
plt.show()
```

```python
import timeit 

qbT2 = []

for i in np.arange(25) : #se probaran con los primeros 25 elementos
    ti = timeit.default_timer()
    dview.apply_sync(primosQB, 2**i)
    qbT2.append(timeit.default_timer() - ti)
    
import matplotlib.pyplot as plt
plt.plot(qbT2, label = "Resultado obtenidos con apply_sync")
plt.legend(loc="best")
plt.show()
```

### Comparando tiempos de primosQB

```python
#Diferencia de tiempos
import matplotlib.pyplot as plt
plt.plot(qbT1, label = "Resultado obtenidos con map_sync")
plt.plot(qbT2, label = "Resultado obtenidos con apply_sync")
plt.legend(loc="best")
plt.show()
```

## <font color='blue'>Funci√≥n primosQC</font>

```python
def primosQC(x):
    import numpy as np
    lista = np.arange(np.int(x**0.5)+1)[2:]
    residuos = np.mod(x, lista)
    return np.all(residuos)
```

```python
import timeit 

qcT1 = []

for i in np.arange(25) : #se probaran con los primeros 25 elementos
    ti = timeit.default_timer()
    dview.map_sync(primosQC, range(25))
    qcT1.append(timeit.default_timer() - ti)
    
import matplotlib.pyplot as plt
plt.plot(qcT1, label = "Resultado obtenidos con map_sync")
plt.legend(loc="best")
plt.show()
```

```python
import timeit 

qcT2 = []

for i in np.arange(25) : #se probaran con los primeros 25 elementos
    ti = timeit.default_timer()
    dview.apply_sync(primosQC, 2**i)
    qcT2.append(timeit.default_timer() - ti)
    
import matplotlib.pyplot as plt
plt.plot(qcT2, label = "Resultado obtenidos con apply_sync")
plt.legend(loc="best")
plt.show()
```

### Comparando tiempos de primosQB

```python
#Diferencia de tiempos
import matplotlib.pyplot as plt
plt.plot(qcT1, label = "Resultado obtenidos con map_sync")
plt.plot(qcT2, label = "Resultado obtenidos con apply_sync")
plt.legend(loc="best")
plt.show()
```
